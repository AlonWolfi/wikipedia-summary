```python
def get_experiment_fold(model_name, fold):
    warnings.filterwarnings("ignore")
    experiment = read_data(get_file_path('experiment.pickle', ['questions_model',model_name, str(fold)]))
    if experiment is None:
        experiment = read_data(get_file_path(model_name + '.pickle', 'question_model'))
#     warnings.filterwarnings("default")
    return experiment
```

```python
def parse_scores(scores):
    scores = pd.DataFrame(scores)
    scores_t = pd.DataFrame(scores).T
    diff_df =pd.DataFrame([])
    for col in scores_t.columns:
        if col == 'y_pred':
            continue
        else:
            scores_t[f'{col}_diff'] = scores_t[col] - scores_t['y_pred']
    scores = scores_t.T
    scores['mean'] = pd.DataFrame(scores).mean(axis=1)
    scores['std'] = pd.DataFrame(scores).std(axis=1) / np.sqrt(5)
    return scores
```

```python
model_name = 'logistic'
scores = {}
for fold in range(5):
    experiment = get_experiment_fold(model_name, fold)
    if experiment is None:
        continue
    fold_params = experiment['params']
    fold_results = experiment['results']
    fold_scores = experiment['scores']
    for s, _scores in fold_scores.items():
        if s not in scores:
            scores[s] = {}
        scores[s][str(fold)] = _scores
        
for s, _scores in scores.items():
    scores[s] = parse_scores(_scores)
    scores[s].columns = [f'{s}_{col}' for col in scores[s].columns]
    print(s)
    display(scores[s])
```
