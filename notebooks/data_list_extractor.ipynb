```python
def remove_first_line_nonesense_xml(path):
    with open(path, 'r', encoding="utf8") as xml_file:
        first_line = xml_file.readline()
        reminder = ''.join(xml_file.readlines())
        xml_parts = first_line.split(' ')
        xml_opening = xml_parts[:1]
        xml_ending = xml_parts[-2:]
        first_line_edited = ' '.join(xml_opening + xml_ending)

    with open(path, 'w', encoding="utf8") as xml_file:
        xml_file.write(first_line_edited)
        xml_file.write(reminder)

remove_first_line_nonesense_xml(data_list_xml_path)
```

```python
lines = page_text.split('\n')
```

```python
def get_lines_generator(lines):
    for line in lines:
        yield line
```

```python
def is_line_main_title(line):
    return (line[:2] == '==' and line[-2:] == '==') and (line[:3] != '===' and line[-3:] != '===')
```

```python
def get_intance_name(line):
    try:
        possible_names = re.search('\[\[([^\]]*)\]\]', line).group(1)
        return possible_names.split('|')[0]
    except:
        return None
```

```python
lines_generator = get_lines_generator(lines)

data_list = list()

line = next(lines_generator)

while not is_line_main_title(line):
    line = next(lines_generator)

line = next(lines_generator)

while not is_line_main_title(line):
    if len(line) > 0 and line[0] == '*':
        instance_name = get_intance_name(line)
        if instance_name:
            data_list.append(instance_name)
    line = next(lines_generator)
data_list
```
